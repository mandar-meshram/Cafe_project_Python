```python
class Bill:
    bill = 0
    items = []

    def addItem(self, item, price):
        self.bill += price
        self.items.append(f"{item}: ${price}")

    def tBill(self, opt):
        print("\nYour Bill Summary:")
        for item in self.items:
            print(item)
        print(f"Total Bill: ${self.bill}\n")

class Dessert:
    def DessertMenu(self, opt):
        print("Press 1 for Ice Cream ($5)")
        print("Press 2 for Butter Milk ($3)")
        print("Press 3 for Gulab Jamun ($4)")
        print("Press 4 for previous menu")

class NonVeg:
    def NonVegMenu(self, opt):
        print("Press 1 for Biryani ($10)")
        print("Press 2 for Butter Chicken ($12)")
        print("Press 3 for Chicken Tikka ($8)")
        print("Press 4 for previous menu")

class Veg:
    def VegMenu(self, opt):
        print("Press 1 for Palak Paneer ($7)")
        print("Press 2 for Butter Paneer ($8)")
        print("Press 3 for Paneer Tikka ($6)")
        print("Press 4 for previous menu")

class Menu:
    def MenuList(self):
        print("Press 1 for Veg Menu")
        print("Press 2 for Non-veg Menu")
        print("Press 3 for Dessert Menu")
        print("Press 4 for Bill")

class Manager:
    def Order(self):
        m = Menu()
        v = Veg()
        n = NonVeg()
        d = Dessert()
        b = Bill()
        flag = True

        while flag:
            m.MenuList()
            opt = int(input("Please order: "))
            if opt == 1:
                while True:
                    v.VegMenu(opt)
                    dish = int(input("Select from the given menu: "))
                    if dish == 4:
                        break
                    elif dish == 1:
                        b.addItem("Palak Paneer", 7)
                        print("Your Veg order is ready")
                    elif dish == 2:
                        b.addItem("Butter Paneer", 8)
                        print("Your Veg order is ready")
                    elif dish == 3:
                        b.addItem("Paneer Tikka", 6)
                        print("Your Veg order is ready")
                    else:
                        print("Invalid option. Try again.")

            elif opt == 2:
                while True:
                    n.NonVegMenu(opt)
                    dish = int(input("Select from the given menu: "))
                    if dish == 4:
                        break
                    elif dish == 1:
                        b.addItem("Biryani", 10)
                        print("Your Non-Veg order is ready")
                    elif dish == 2:
                        b.addItem("Butter Chicken", 12)
                        print("Your Non-Veg order is ready")
                    elif dish == 3:
                        b.addItem("Chicken Tikka", 8)
                        print("Your Non-Veg order is ready")
                    else:
                        print("Invalid option. Try again.")

            elif opt == 3:
                while True:
                    d.DessertMenu(opt)
                    dish = int(input("Select from the given menu: "))
                    if dish == 4:
                        break
                    elif dish == 1:
                        b.addItem("Ice Cream", 5)
                        print("Your Dessert order is ready")
                    elif dish == 2:
                        b.addItem("Butter Milk", 3)
                        print("Your Dessert order is ready")
                    elif dish == 3:
                        b.addItem("Gulab Jamun", 4)
                        print("Your Dessert order is ready")
                    else:
                        print("Invalid option. Try again.")

            elif opt == 4:
                b.tBill(opt)
                flag = False
            else:
                print("Please select a valid option.")

class Hotel:
    def welcome(self):
        print("Welcome to Raddison Blue Hotel")
        m = Manager()
        m.Order()

#******START*****
h = Hotel()
h.welcome()
```
